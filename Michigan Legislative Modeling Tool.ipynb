```python
# Sets a list that will be changed and updated to the measured conditions through the current year. 
workingList = CO2EmissionsHardCoded2[:]

# Looping from the current year through 2050 
for i in range(currentYear,2051):
    # Each year, the yearly change is first set to the value of the business as usual conditions. 
    yearlyChange = CO2EmissionsBAUYOYChange[i-2024]
    # For each reduction measure that is turned on, the resulting yearly change from that reduction measure being added to 
    # the total yearly impact. 
    if RM1 == True:
        yearlyChange += CO2EmissionsElectricityProjectionsYOYChangeRM1[i-2024]
    if RM2 == True:
        yearlyChange += CO2EmissionsProjectionsYOYChangeRM2[i-2024]
    if RM3 == True:
        yearlyChange += CO2EmissionsProjectionsYOYChangeRM3[i-2024]
    if RM4 == True: 
        yearlyChange += CO2EmissionsProjectionsYOYChangeRM4[i-2024]
    if RM5 == True: 
        yearlyChange += CO2EmissionsProjectionsYOYChangeRM5[i-2024]
    if RM6 == True: 
        yearlyChange += CO2EmissionsProjectionsYOYChangeRM6[i-2024]
    if RM7 == True: 
        yearlyChange += CO2EmissionsProjectionsYOYChangeRM7[i-2024]
    if RM8 == True: 
        yearlyChange += CO2EmissionsProjectionsYOYChangeRM8[i-2024]
    if RM9 == True: 
        yearlyChange += CO2EmissionsProjectionsYOYChangeRM9[i-2024]
    if RM10 == True: 
        yearlyChange += CO2EmissionsProjectionsYOYChangeRM10[i-2024]

    # These lines locate the previous years value, add the yearly change, and then append it to the working list. 
    previous = workingList[-1]
    yearlyValue = previous + yearlyChange
    workingList.append(yearlyValue)

# Uncomment the below line in order to see the yearly values of CO2e emitted each year.
#print(workingList)

# Plotting both the business as usual case as a baseline and the input from the active reduction measures
plt.figure(figsize=(11,6))
plt.plot(years, CO2EmissionsHardCoded2+CO2EmissionsBAU, color="orange", label="Business As Usual")
plt.plot(years, workingList, color="royalblue", label="Custom Input")
plt.xlabel("Year")
plt.ylabel("CO2 Equivalent Emissions (MMT)")
plt.title("Pre-Existing Reduction Measure Impacts")
plt.ylim(0,200)
plt.grid()
plt.legend()
```

## Pre-Existing Reduction Measure Tool - Split Goals

#### Set Parameters

Set any of the variables below to "True" to turn them on, and "False" to turn them off. These variables correspond to milestone goals set in each reduction measure. Some have goals in 2030, 2050, or both. If a Reduction Measure has a goal in 2030 and 2050, both must be turned on in order for the tool to take the 2050 impact into consideration. If 2030 is off, then neither 2030 nor 2050 will not be calculated for that reduction measure. 

```python
RM1Electricity2030 = False
RM1Electricity2050 = False

RM22030 = False

RM32050 = False

RM42050 = False

RM52030 = False
RM52050 = False

RM62030 = False
RM62050 = False

RM72030 = False
RM72050 = False

RM82030 = False

RM92030 = False
RM92050 = False

RM102030 = False
RM102050 = False
```

#### Calculate Year By Year Changes

There are no inputs in the cell below. After running the reduction measure cell above, run the cell below to output a graph displaying the reduction measure impact.

```python
# Sets a list that will be changed and updated to the measured conditions through the current year. 
workingListSplit = CO2EmissionsHardCoded2[:]

# Looping from the current year through 2050 
for i in range(currentYear,2051):
    # Each year, the yearly change is first set to the value of the business as usual conditions.
    yearlyChange = CO2EmissionsBAUYOYChange[i-2024]
    # Since any split measures track their milestones between 2030 and 2050, this code interpolates between the current year 
    # and 2030, then 2031 and 2050 in the next section. 
    if i <= 2030:
        # For each reduction measure split starting in 2030 that is turned on, the resulting yearly change from that reduction 
        # measure being added to the total yearly impact. 
        if RM1Electricity2030 == True:
            yearlyChange += CO2EmissionsElectricityProjectionsYOYChangeRM1[i-2024]
        if RM22030 == True:
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM2[i-2024]
        if RM32050 == True:
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM3[i-2024]
        if RM42050 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM4[i-2024]
        if RM52030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM5[i-2024]
        if RM62030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM6[i-2024]
        if RM72030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM7[i-2024]
        if RM82030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM8[i-2024]
        if RM92030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM9[i-2024]
        if RM102030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM10[i-2024]
            
        # These lines locate the previous years value, add the yearly change, and then append it to the working list.
        previous = workingListSplit[-1]
        yearlyValue = previous + yearlyChange
        workingListSplit.append(yearlyValue)
        
    elif i > 2030:
        # For each reduction measure split starting in 2031 that is turned on, the resulting yearly change from that reduction 
        # measure being added to the total yearly impact. 
        if RM1Electricity2050 == True and RM1Electricity2030 == True:
            yearlyChange += CO2EmissionsElectricityProjectionsYOYChangeRM1[i-2024]
        if RM22030 == True:
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM2[i-2024]
        if RM32050 == True:
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM3[i-2024]
        if RM42050 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM4[i-2024]
        if RM52050 == True and RM52030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM5[i-2024]
        if RM62050 == True and RM62030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM6[i-2024]
        if RM72050 == True and RM72030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM7[i-2024]
        if RM82030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM8[i-2024]
        if RM92050 == True and RM92030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM9[i-2024]
        if RM102050 == True and RM102030 == True: 
            yearlyChange += CO2EmissionsProjectionsYOYChangeRM10[i-2024]
            
        # These lines locate the previous years value, add the yearly change, and then append it to the working list.
        previous = workingListSplit[-1]
        yearlyValue = previous + yearlyChange
        workingListSplit.append(yearlyValue)

# Uncomment the below line in order to see the yearly values of CO2e emitted each year.
#print(workingListSplit)

# Plotting both the business as usual case as a baseline and the input from the active reduction measures
plt.figure(figsize=(11,6))
plt.plot(years, CO2EmissionsHardCoded2+CO2EmissionsBAU, color="orange", label="Business As Usual")
plt.plot(years, workingListSplit, color="royalblue", label="Custom Input")
plt.xlabel("Year")
plt.ylabel("CO2 Equivalent Emissions (MMT)")
plt.title("Pre-Existing Reduction Measure Impacts - Split Goals")
plt.ylim(0,200)
plt.grid()
plt.legend()
```

### Custom Reduction Measure Function

```python
# Function designed to take in a single custom reduction measure and output a total CO2e reduction from that measure.
def customReductionMeasure(reductionMeasure, year, percent):
    # If statements used to determine which reduction measure the custom input is dealing with
    if reductionMeasure == 1:
        customYOYPercents = RM1Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM1.copy()
        RM1 = False
    elif reductionMeasure == 2:
        customYOYPercents = RM2Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM2.copy()
        RM2 = False
    elif reductionMeasure == 3:
        customYOYPercents = RM3Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM3.copy()
        RM3 = False
    elif reductionMeasure == 4:
        customYOYPercents = RM4Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM4.copy()
        RM4 = False
    elif reductionMeasure == 5:
        customYOYPercents = RM5Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM5.copy()
        RM5 = False
    elif reductionMeasure == 6:
        customYOYPercents = RM6Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM6.copy()
        RM6 = False
    elif reductionMeasure == 7:
        customYOYPercents = RM7Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM7.copy()
        RM7 = False
    elif reductionMeasure == 8:
        customYOYPercents = RM8Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM8.copy()
        RM8 = False
    elif reductionMeasure == 9:
        customYOYPercents = RM9Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM9.copy()
        RM9 = False
    elif reductionMeasure == 10:
        customYOYPercents = RM10Percentages.copy()
        customYOYReductions = CO2EmissionsProjectionsYOYChangeRM10.copy()
        RM10 = False
    
    # If the input percent is less than the maximum percentage in the RM but greather than the lowest value, 
    # find where it is in the percentages and calculate the total reduction. 
    totalReduction = 0
    if percent <= customYOYPercents[-1] and percent >= customYOYPercents[0]:
        timeyear = 2024
        for i in customYOYPercents:
            if percent <= i:
                break
            else: 
                timeyear += 1
        for i in range(2024,timeyear):
            totalReduction += customYOYReductions[i-2024]
        percentOfStep = (percent - customYOYPercents[timeyear-2024-1]) / (customYOYPercents[timeyear-2024] - 
                                                                          customYOYPercents[timeyear-2024-1])
        totalReduction += (customYOYReductions[timeyear-2024-1] * percentOfStep)
        
    # If the input percent is less than the lowest value, scale down the lowest value 
    if percent < customYOYPercents[0]:
        if reductionMeasure == 1:
            percentOfStep = percent/(customYOYPercents[0]-16)
            totalReduction = customYOYReductions[0] * percentOfStep
        else:
            percentOfStep = percent/customYOYPercents[0]
            totalReduction = customYOYReductions[0] * percentOfStep
        
    # If the input percent is more than the last value, scale up the sum of the reduction within the measure. 
    if percent > customYOYPercents[-1]:
        sumOfMeasure = sum(customYOYReductions)
        totalReduction = sumOfMeasure * (percent/customYOYPercents[-1])
    
    reductionLog(reductionMeasure,year,percent,totalReduction)
    return(totalReduction)
```
